```python
class menu:

    def __init__(self):
        self.root=Tk()
        self.root.title('Menu') #title of window to be displayed
        self.root.state('zoomed')
        conn=sqlite3.connect('test3.db') #opens database(db) or creates one
        
        #creating BOOK and MEMBER table
        conn.execute('''create table if not exists BOOK  
        (ID VARCHAR PRIMARY KEY NOT NULL,
        TITLE VARTEXT NOT NULL,
        AUTHOR VARTEXT NOT NULL,
        GENRE VARTEXT NOT NULL,
        COPIES VARINT NOT NULL,
        LOCATION VARCHAR NOT NULL);''')
        conn.commit() #saves changes made in db
        conn.execute('''create table if not exists MEMBER
        (BOOK_ID VARCHAR NOT NULL,
        MEMBER_ID VARCHAR NOT NULL,
        ISSUE_DATE DATE NOT NULL,
        RETURN_DATE DATE NOT NULL,
        PRIMARY KEY (BOOK_ID,MEMBER_ID));''') 
        conn.commit()
        
        #adding values to book table. [After one time executing comment below line as duplicate entries will throw error on re-run)
        #conn=sqlite3.connect('test3.db')
        #conn.execute("INSERT INTO BOOK (ID,TITLE,AUTHOR,GENRE,COPIES,LOCATION) \
        #VALUES ('B788','The Lean Startup','Eric Ries','business',1,'s1')"); # '\' means new line
        #conn.execute("INSERT INTO BOOK (ID,TITLE,AUTHOR,GENRE,COPIES,LOCATION) \
        #VALUES ('B110','The 7 Habits','Stephen Covey','business',3,'s2')"); 
        #conn.execute("INSERT INTO BOOK (ID,TITLE,AUTHOR,GENRE,COPIES,LOCATION) \
        #VALUES ('F134','The Alchemist','Paulo Coelho','fiction',2,'s3')");
        
        
        #adding values to member table (Once data is loaded in database comment below line as duplicate entries will throw error on re-run)
        #conn.execute("INSERT INTO MEMBER (BOOK_ID,MEMBER_ID,ISSUE_DATE,RETURN_DATE) \
        #VALUES (0,'U1',0,0)");
        #conn.execute("INSERT INTO MEMBER (BOOK_ID,MEMBER_ID,ISSUE_DATE,RETURN_DATE) \
        #VALUES (0,'A1',0,0)");
        #conn.execute("INSERT INTO MEMBER (BOOK_ID,MEMBER_ID,ISSUE_DATE,RETURN_DATE) \
        #VALUES (0,'J1',0,0)");
        #conn.commit()
        #conn.close()#closes db
        
        self.a=self.canvases(image1) #calls method to insert background image to the page.'text'=text to be dispalayed on button. bg=background colour. padx=padding bodder depth
        l1=Button(self.a,text='BOOK DATA',font='Papyrus 22 bold',fg='Yellow',bg='Black',width=19,padx=10,borderwidth=0,command=self.book).place(x=100,y=500) #'command'calls the method for the action to be performed on clicking the button
        l2=Button(self.a,text='MEMBER DATA',font='Papyrus 22 bold',fg='Yellow',bg='Black',width=19,padx=10,borderwidth=0,command=self.student).place(x=800,y=500)
        self.root.mainloop() 
        
        
    def canvases(self,images): #used for setting image for GUI window
        w = self.root.winfo_screenwidth()
        h = self.root.winfo_screenheight()
       
        photo=Image.open(images)
        photo1=photo.resize((w,h),Image.ANTIALIAS) 
        photo2=ImageTk.PhotoImage(photo1) 

      
        self.canvas = Canvas(self.root, width='%d'%w, height='%d'%h) 
        self.canvas.grid(row = 0, column = 0) #grid-mentions position
        self.canvas.grid_propagate(0) 
        self.canvas.create_image(0, 0, anchor = NW, image=photo2) # NW-northwest position
        self.canvas.image=photo2 
        return self.canvas 
    
    def book(self): #options available in bookpage
        self.a.destroy() 
        self.a=self.canvases(image2)
        l1=Button(self.a,text='Add Books',font='Papyrus 22 bold',fg='Orange',bg='Black',width=15,padx=10,command=self.addbook).place(x=12,y=100) #to ensure proper allignment of labels their x axis is given same value and y axis having value difference of 100 so that labels are placed one below other.
        l2=Button(self.a,text='Search Books',font='Papyrus 22 bold',fg='Orange',bg='Black',width=15,padx=10,command=self.search).place(x=12,y=200)

        l4=Button(self.a,text='All Books',font='Papyrus 22 bold',fg='Orange',bg='Black',width=15,padx=10,command=self.all).place(x=12,y=300)
        l4=Button(self.a,text='<< Main Menu',font='Papyrus 22 bold',fg='Orange',bg='Black',width=15,padx=10,command=self.mainmenu).place(x=12,y=500)



    def addbook(self):
        self.aid=StringVar() #takes string input(bookid) from user
        self.aauthor=StringVar() #authour name
        self.aname=StringVar() #book name
        self.acopies=IntVar() #current no of copies of the book
        self.agenre=StringVar()#genre of the book to be added
        self.aloc=StringVar()#shelf rack where book to be arranged
        self.f1=Frame(self.a,height=500,width=650,bg='black') #frame dimension
        self.f1.place(x=500,y=100) #frame position
        l1=Label(self.f1,text='Book ID : ',font='Papyrus 12 bold',fg='Orange',bg='Black',pady=1).place(x=50,y=50)
        e1=Entry(self.f1,width=45,bg='orange',fg='black',textvariable=self.aid).place(x=150,y=50) #stores user input in a variable
        l2=Label(self.f1,text='Title : ',font='Papyrus 12 bold',fg='Orange',bg='Black',pady=1).place(x=50,y=100)
        e2=Entry(self.f1,width=45,bg='orange',fg='black',textvariable=self.aname).place(x=150,y=100)
        l3=Label(self.f1,text='Author : ',font='Papyrus 12 bold',fg='orange',bg='Black',pady=1).place(x=50,y=150)
        e3=Entry(self.f1,width=45,bg='orange',fg='black',textvariable=self.aauthor).place(x=150,y=150)
        l4=Label(self.f1,text='Genre : ',font='Papyrus 12 bold',fg='orange',bg='Black',pady=1).place(x=50,y=200)
        e2=Entry(self.f1,width=45,bg='orange',fg='black',textvariable=self.agenre).place(x=150,y=200)
        l4=Label(self.f1,text='Copies : ',font='Papyrus 12 bold',fg='orange',bg='Black',pady=1).place(x=50,y=250)
        e2=Entry(self.f1,width=45,bg='orange',fg='black',textvariable=self.acopies).place(x=150,y=250)
        l5=Label(self.f1,text='Location : ',font='Papyrus 12 bold',fg='orange',bg='Black',pady=1).place(x=50,y=300)
        e3=Entry(self.f1,width=45,bg='orange',fg='black',textvariable=self.aloc).place(x=150,y=300)
        self.f1.grid_propagate(0)
        b1=Button(self.f1,text='Add',font='Papyrus 10 bold',fg='black',bg='orange',width=15,bd=3,command=self.adddata).place(x=150,y=400)
        b2=Button(self.f1,text='Back',font='Papyrus 10 bold',fg='black',bg='orange',width=15,bd=3,command=self.rm).place(x=350,y=400)

    def rm(self):
        self.f1.destroy() #closes particular window
    def mainmenu(self):
        self.root.destroy()
        a=menu() #calls main menu

    def adddata(self):
        a=self.aid.get() #for bookid
        b=self.aname.get() #for bookname
        c=self.aauthor.get()#for author name
        d=self.agenre.get() #for genre of book
        e=self.acopies.get() #for number of copies of book purchased
        f=self.aloc.get()
        conn=sqlite3.connect('test3.db')
        try:
            if (a and b and c and d  and f)=="":
                messagebox.showinfo("Error","Fields cannot be empty.")
            else:
                conn.execute("insert into BOOK \
                values (?,?,?,?,?,?)",(a.capitalize(),b.capitalize(),c.capitalize(),d.capitalize(),e,f.capitalize(),)); #not case sensitive for input by user
                conn.commit()
                messagebox.showinfo("Success","Book added successfully")
        except sqlite3.IntegrityError:
            messagebox.showinfo("Error","Book is already present.")
        conn.close()

    def search(self):
   
        self.sid=StringVar()
        self.f1=Frame(self.a,height=500,width=650,bg='black')
        self.f1.place(x=500,y=100)
        l1=Label(self.f1,text='Book ID: ',font=('Papyrus 10 bold'),bd=2, fg='orange',bg='black').place(x=20,y=40)
        e1=Entry(self.f1,width=25,bd=5,bg='orange',fg='black',textvariable=self.sid).place(x=260,y=40)
        b1=Button(self.f1,text='Search',bg='orange',font='Papyrus 10 bold',width=9,bd=2,command=self.serch1).place(x=500,y=37)
        b1=Button(self.f1,text='Back',bg='orange',font='Papyrus 10 bold',width=10,bd=2,command=self.rm).place(x=250,y=450)

    def create_tree(self,plc,lists):
        self.tree=ttk.Treeview(plc,height=13,column=(lists),show='headings') #to display details in treeview
        n=0
        while n is not len(lists):
            self.tree.heading("#"+str(n+1),text=lists[n])
            self.tree.column(""+lists[n],width=100)
            n=n+1
        return self.tree


    def serch1(self):
        k=self.sid.get() #gets value stored in the variable
        if k!="":
            self.list4=("BOOK ID","TITLE","AUTHOR","GENRE","COPIES","LOCATION") #Treeview heading of columns
            self.trees=self.create_tree(self.f1,self.list4) 
            self.trees.place(x=25,y=150)
            conn=sqlite3.connect('test3.db')

            c=conn.execute("select * from BOOK where ID=? OR TITLE=? OR AUTHOR=? OR GENRE=?",(k.capitalize(),k.capitalize(),k.capitalize(),k.capitalize(),))
            a=c.fetchall() #fetch all records meeting the condition
            if len(a)!=0:
                for row in a:
                    self.trees.insert("",END,values=row)  
                conn.commit()
                conn.close()
                self.trees.bind('<<TreeviewSelect>>')  
                self.variable = StringVar(self.f1)    #f1 store frame size details
                self.variable.set("Select Action:")  


                self.cm =ttk.Combobox(self.f1,textvariable=self.variable ,state='readonly',font='Papyrus 15 bold',height=50,width=15,) #
                self.cm.config(values =('Add Copies', 'Delete Copies', 'Delete Book')) 

                self.cm.place(x=50,y=100)
                self.cm.pack_propagate(0)


                self.cm.bind("<<ComboboxSelected>>",self.combo)
                self.cm.selection_clear()
            else:
                messagebox.showinfo("Error","Data not found")



        else:
            messagebox.showinfo("Error","Search field cannot be empty.")


    def combo(self,event): 
        self.var_Selected = self.cm.current() 
        if self.var_Selected==0:
            self.copies(self.var_Selected) 
        elif self.var_Selected==1:
            self.copies(self.var_Selected)
        elif self.var_Selected==2:
            self.deleteitem()
    def deleteitem(self):
        try:
            self.curItem = self.trees.focus()

            self.c1=self.trees.item(self.curItem,"values")[0]
            b1=Button(self.f1,text='Update',font='Papyrus 10 bold',width=9,bd=3,command=self.delete2).place(x=500,y=97)

        except:
            messagebox.showinfo("Empty","Please select something.")
    def delete2(self): 
        conn=sqlite3.connect('test3.db')
        cd=conn.execute("select * from MEMBER where BOOK_ID=?",(self.c1,))
        ab=cd.fetchall()
        if ab!=0:
            conn.execute("DELETE FROM BOOK where ID=?",(self.c1,));
            conn.commit()
            messagebox.showinfo("Successful","Book Deleted sucessfully.")
            self.trees.delete(self.curItem)
        else:
            messagebox.showinfo("Error","Book is Issued.\nBook cannot be deleted.")
        conn.commit()
        conn.close()


    def copies(self,varr): 
        try:
            curItem = self.trees.focus() 
            self.c1=self.trees.item(curItem,"values")[0] 
            self.c2=self.trees.item(curItem,"values")[4] 
            self.scop=IntVar() #
            self.e5=Entry(self.f1,width=20,textvariable=self.scop) 
            self.e5.place(x=310,y=100)
            if varr==0: #
                b5=Button(self.f1,text='Update',font='Papyrus 10 bold',bg='orange',fg='black',width=9,bd=3,command=self.copiesadd).place(x=500,y=97)
            if varr==1: #
                b6=Button(self.f1,text='Update',font='Papyrus 10 bold',bg='orange',fg='black',width=9,bd=3,command=self.copiesdelete).place(x=500,y=97)
        except:
            messagebox.showinfo("Empty","Please select something.")

    def copiesadd(self):
        no=self.e5.get()
        if int(no)>=0:

            conn=sqlite3.connect('test3.db')

            conn.execute("update BOOK set COPIES=COPIES+? where ID=?",(no,self.c1,)) #'?'captures user input on number of book copies to be added
            conn.commit()

            messagebox.showinfo("Updated","Copies added sucessfully.")
            self.serch1()
            conn.close()

        else:
            messagebox.showinfo("Error","No. of copies cannot be negative.")

    def copiesdelete(self):
        no1=self.e5.get()
        if int(no1)>=0:
            if int(no1)<=int(self.c2): 
                conn=sqlite3.connect('test3.db')

                conn.execute("update BOOK set COPIES=COPIES-? where ID=?",(no1,self.c1,)) 
                conn.commit() 
                conn.close()

                messagebox.showinfo("Updated","Deleted sucessfully")
                self.serch1()

            else:
                messagebox.showinfo("Maximum","No. of copies to delete exceed available copies.")
        else:
            messagebox.showinfo("Error","No. of copies cannot be negative.")

    def all(self):
        self.f1=Frame(self.a,height=500,width=650,bg='black')
        self.f1.place(x=500,y=100)
        b1=Button(self.f1,text='Back',bg='orange' ,fg='black',width=10,bd=3,command=self.rm).place(x=250,y=400)
        conn=sqlite3.connect('test3.db')
        self.list3=("BOOK ID","TITLE","AUTHOR","GENRE","COPIES","LOCATION")
        self.treess=self.create_tree(self.f1,self.list3)
        self.treess.place(x=25,y=50)
        c=conn.execute("select * from BOOK")
        g=c.fetchall()
        if len(g)!=0:
            for row in g:
                self.treess.insert('',END,values=row)
        conn.commit()
        conn.close()

    def student(self):
        self.a.destroy() 
        self.a=self.canvases(image2)
        l1=Button(self.a,text='Issue book',font='Papyrus 22 bold',fg='Orange',bg='Black',width=15,padx=10,command=self.issue).place(x=12,y=100)
        l2=Button(self.a,text='Return Book',font='Papyrus 22 bold',fg='Orange',bg='Black',width=15,padx=10,command=self.returnn).place(x=12,y=200)
        l3=Button(self.a,text='Member Details',font='Papyrus 22 bold',fg='Orange',bg='Black',width=15,padx=10,command=self.activity).place(x=12,y=300)
        l4=Button(self.a,text='<< Main Menu',font='Papyrus 22 bold',fg='Orange',bg='Black',width=15,padx=10,command=self.mainmenu).place(x=12,y=600)
        l5=Button(self.a,text='Register User',font='Papyrus 22 bold',fg='Orange',bg='Black',width=15,padx=10,command=self.register).place(x=12,y=400)
        
    def register(self):
        self.astudentt=StringVar() #to capture memberid typed by user
        self.f1=Frame(self.a,height=550,width=500,bg='black') #frame dimensions foe member page
        self.f1.place(x=500,y=100)
        l2=Label(self.f1,text='Member Id : ',font='papyrus 15 bold',bg='black',fg='orange').place(x=50,y=150)
        e2=Entry(self.f1,width=25,bd=4,bg='orange',textvariable=self.astudentt).place(x=180,y=150)
        b1=Button(self.f1,text='Back',font='Papyrus 10 bold',fg='black',bg='orange',width=10,bd=3,command=self.rm).place(x=50,y=250)
        b1=Button(self.f1,text='Register',font='Papyrus 10 bold',fg='black',bg='orange',width=10,bd=3,command=self.adduser).place(x=200,y=250)    

        
    def adduser(self):
        b=self.astudentt.get() #to get stored studentid captured from user input
        conn=sqlite3.connect('test3.db') #opening db
        try:
            if (b)=="":
                messagebox.showinfo("Error","Fields cannot be empty.")
            else:
                conn.execute("insert into MEMBER \
                values (0,?,0,0)",(b.capitalize(),));
                conn.commit()
                messagebox.showinfo("Success","Registration successfull")
        except sqlite3.IntegrityError:
            messagebox.showinfo("Error","User is existing") #due to primary key constraint of stuid act as check for duplicate user
        conn.close()

       

    def issue(self):
        self.aidd=StringVar() #book id to be issued by user
        self.astudentt=StringVar() #member id of the member issuing book
        self.f1=Frame(self.a,height=550,width=500,bg='black')
        self.f1.place(x=500,y=100)
        l1=Label(self.f1,text='Book ID : ',font='papyrus 15 bold',bg='black',fg='orange').place(x=50,y=100)
        e1=Entry(self.f1,width=25,bd=4,bg='orange',textvariable=self.aidd).place(x=180,y=100)
        l2=Label(self.f1,text='Member Id : ',font='papyrus 15 bold',bg='black',fg='orange').place(x=50,y=150)
        e2=Entry(self.f1,width=25,bd=4,bg='orange',textvariable=self.astudentt).place(x=180,y=150)
        b1=Button(self.f1,text='Back',font='Papyrus 10 bold',fg='black',bg='orange',width=10,bd=3,command=self.rm).place(x=50,y=250)
        b1=Button(self.f1,text='Issue',font='Papyrus 10 bold',fg='black',bg='orange',width=10,bd=3,command=self.issuedbook).place(x=200,y=250)
           
    
    def issuedbook(self):
        bookid=self.aidd.get()
        studentid=self.astudentt.get()
        conn=sqlite3.connect('test3.db')
        cursor=conn.cursor()
        cursor.execute("select ID,COPIES from BOOK where ID=?",(bookid.capitalize(),))
        an=cursor.fetchall()
        cursor2=conn.cursor()
        cursor2.execute("select BOOK_ID,MEMBER_ID from MEMBER where MEMBER_ID=?",(studentid.capitalize(),))
        ab=cursor2.fetchall()       
        if (bookid and studentid!=""): #check whether any records in db matches condition
            if ab!=[]:
                if an!=[]: #checks presence of particular book
                    for i in an:
                        if i[1]>0:
                            try:
                                conn.execute("insert into MEMBER \
                                values (?,?,date('now'),date('now','+7 day'))",(bookid.capitalize(),studentid.capitalize(),)); #user can hold book for 7 days for usual subscription. 'now' captures current date
                                conn.execute("update BOOK set COPIES=COPIES-1 where ID=?",(bookid.capitalize(),)) #updates current number of book copies in BOOK table
                                conn.commit()
                                conn.close()
                                messagebox.showinfo("Updated","Book Issued sucessfully.")
                            except:
                                messagebox.showinfo("Error","Book is already issued by member.")
                        else:
                            messagebox.showinfo("Unavailable","Book unavailable.\nThere are 0 copies of the book.")
                else:
                    messagebox.showinfo("Error","No such Book in Database.")
            else:
                messagebox.showinfo("Error","Invalid user,please register.")
        else:
            messagebox.showinfo("Error","Fields cannot be blank.")
                
                   

    def returnn(self):
        self.aidd=StringVar()
        self.astudentt=StringVar()
        self.astype=StringVar() #type of member subscription(stu=student, fac=faculty)
        self.f1=Frame(self.a,height=550,width=500,bg='black')
        self.f1.place(x=500,y=100)
        l1=Label(self.f1,text='Book ID : ',font='papyrus 15 bold',fg='orange', bg='black').place(x=50,y=100)
        e1=Entry(self.f1,width=25,bd=4,bg='orange',textvariable=self.aidd).place(x=190,y=100)
        l2=Label(self.f1,text='Member Id : ',font='papyrus 15 bold',fg='orange', bg='black').place(x=50,y=150)
        e2=Entry(self.f1,width=25,bd=4,bg='orange',textvariable=self.astudentt).place(x=190,y=150)
        l3=Label(self.f1,text='Member type:',font='papyrus 15 bold',fg='orange', bg='black').place(x=50,y=200)
        e3=Entry(self.f1,width=25,bd=4,bg='orange',textvariable=self.astype).place(x=190,y=200)
        b1=Button(self.f1,text='Back',font='Papyrus 10 bold',bg='orange',fg='black',width=10,bd=3,command=self.rm).place(x=50,y=250)
        b1=Button(self.f1,text='Return',font='Papyrus 10 bold',bg='orange',fg='black',width=10,bd=3,command=self.returnbook).place(x=150,y=250) #x=200
        b1=Button(self.f1,text='Fine',font='Papyrus 10 bold',bg='orange',fg='black',width=10,bd=3,command=self.fine).place(x=250,y=250)
        self.f1.grid_propagate(0)


    def fine(self): # press FINE button before pressing RETURN button
        a=self.aidd.get()
        b=self.astudentt.get()
        t=self.astype.get()
        conn=sqlite3.connect('test3.db')
        c=conn.execute("select RETURN_DATE from MEMBER where BOOK_ID=? and MEMBER_ID=?",(a.capitalize(),b.capitalize(),))
        d=c.fetchone()
        conn.commit()
        conn.close()
        a = dt.datetime.strptime(d[0],'%Y-%m-%d') #to convert from string to datetimeformat
        c=a.date()
        g=dt.datetime.now().strftime('%Y-%m-%d') #to convert datetime to string format
        e=dt.datetime.strptime(g,'%Y-%m-%d')
        i=e.date()
        delta = i-c    #calculates number of days between 2 dates
        if(t=='STU' or t=='stu'):
            if (delta.days>7): #student subscription allows 7 days of holding books
                fine = delta.days*5 #fine rate for faculty= Rs.5
            else:
                fine =0.0
            messagebox.showinfo("Fine","Fine to be paid: %s" %(fine)) 
        elif(t=='FAC' or t=='fac'):
            if (delta.days>14): #faculty subscription allows 14 days of holding books
                fine = delta.days*2 #fine rate for faculty= Rs.2
            else:
                fine =0.0
            messagebox.showinfo("Fine","Fine to be paid: %s" %(fine))
        else:
            messagebox.showinfo("Error","Invalid member type")
        
       
                

    def returnbook(self): 
        a=self.aidd.get()
        b=self.astudentt.get()

        conn=sqlite3.connect('test3.db')

        fg=conn.execute("select ID from BOOK where ID=?",(a.capitalize(),))
        fh=fg.fetchall()
        conn.commit()
        if fh!=None:
            c=conn.execute("select * from MEMBER where BOOK_ID=? and MEMBER_ID=?",(a.capitalize(),b.capitalize(),))
            d=c.fetchall()
            conn.commit()
            if len(d)!=0:
                c.execute("DELETE FROM MEMBER where BOOK_ID=? and MEMBER_ID=?",(a.capitalize(),b.capitalize(),));
                conn.commit()
                conn.execute("update BOOK set COPIES=COPIES+1 where ID=?",(a.capitalize(),))
                conn.commit()

                messagebox.showinfo("Success","Book Returned sucessfully.")
            else:
                messagebox.showinfo("Error","Data not found.") #working fine not taking invalid user/mismatch btwn user and bookid
        else:
            messagebox.showinfo("Error","No such book.\nPlease add the book in database.") #\n-new line
        conn.commit()
        conn.close()

    def activity(self): #displays entire member details list
        self.aidd=StringVar()
        self.astudentt=StringVar()
        self.f1=Frame(self.a,height=550,width=500,bg='black')
        self.f1.place(x=500,y=80)
        self.list2=("BOOK ID","MEMBER ID","ISSUE DATE","RETURN DATE")
        self.trees=self.create_tree(self.f1,self.list2)
        self.trees.place(x=50,y=150)


        l1=Label(self.f1,text='Book/Member ID : ',font='Papyrus 15 bold',fg='Orange',bg='black').place(x=50,y=30) 
        e1=Entry(self.f1,width=20,bd=4,bg='orange',textvariable=self.aidd).place(x=280,y=35)
    
        b1=Button(self.f1,text='Back',bg='orange',font='Papyrus 10 bold',width=10,bd=3,command=self.rm).place(x=340,y=450)
        b1=Button(self.f1,text='Search',bg='orange',font='Papyrus 10 bold',width=10,bd=3,command=self.searchact).place(x=40,y=450)
        b1=Button(self.f1,text='All',bg='orange',font='Papyrus 10 bold',width=10,bd=3,command=self.searchall).place(x=190,y=450)
        self.f1.grid_propagate(0)

    def searchact(self): #to search for particular member's activity/status
        self.list2=("BOOK ID","MEMBER ID","ISSUE DATE","RETURN DATE")
        self.trees=self.create_tree(self.f1,self.list2)
        self.trees.place(x=50,y=150)
        conn=sqlite3.connect('test3.db')
        bid=self.aidd.get()
        try:
            c=conn.execute("select * from MEMBER where BOOK_ID=? or MEMBER_ID=?",(bid.capitalize(),bid.capitalize(),))
            d=c.fetchall()
            if len(d)!=0:
                for row in d:
                    self.trees.insert("",END,values=row)
            else:
                messagebox.showinfo("Error","Data not found.")
            conn.commit()

        except Exception as e:
            messagebox.showinfo(e)
        conn.close()

    def searchall(self):
        self.list2=("BOOK ID","MEMBER ID","ISSUE DATE","RETURN DATE")
        self.trees=self.create_tree(self.f1,self.list2)
        self.trees.place(x=50,y=150)
        conn=sqlite3.connect('test3.db')
        try:
            c=conn.execute("select * from MEMBER")
            d=c.fetchall()
            for row in d:
                self.trees.insert("",END,values=row)
            conn.commit()

        except Exception as e:     #handles exception situation without abrupt ending of program
            messagebox.showinfo(e)
        conn.close()
```
